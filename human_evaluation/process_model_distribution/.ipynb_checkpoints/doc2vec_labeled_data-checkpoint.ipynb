```python
tagged_data = [TaggedDocument(process, tags=[str(i)]) 
               for i, process in enumerate(process_df['process'])]

max_epochs = 500
vec_size = 100
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm=1)
  
model.build_vocab(tagged_data)
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch+1))

    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    
    model.alpha -= 0.0002
    model.min_alpha = model.alpha
```

```python
X = np.array([model.dv[str(i)] for i in range(len(tagged_data))])
X_embedded = TSNE(n_components=2, learning_rate='auto',
                   init='random').fit_transform(X)
```

```python
process_doc = data['document']
process_df = pd.DataFrame.from_dict(process_doc)
process_df.head()
```

```python
process_df['process'] = process_df[process_df.columns[:]].apply(
    lambda x: ', '.join(x.dropna().astype(str)),
    axis=1
)
# Remove all columns between column name 'B' to 'D'
process_df.drop(process_df.iloc[:, 0:-1], axis = 1, inplace = True)
process_df['label'] = data['summary']
process_df.head()
```

```python
for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch+1))

    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.epochs)
    
    model.alpha -= 0.0002
    model.min_alpha = model.alpha
```

```python
X = np.array([model.dv[str(i)] for i in range(len(tagged_data))])
X_embedded = TSNE(n_components=2, learning_rate='auto',
                   init='random').fit_transform(X)
```

### Process Model Complexity Distribution

```python
with open('labeled_dataset.json', 'r') as f:
    file_data = json.load(f)
```

```python
file_data['document'].keys()
```

```python
test_idx = [ 58, 140,  51, 295,  25,  54,  82, 274, 222, 204, 305,  45, 291,
            214, 253,  42, 139, 199,  13, 300,  63, 230, 137, 261, 151, 296,
            169, 217, 280, 255, 187, 118, 227,  14, 312, 353,  20, 123, 237,
             95, 307, 359, 209, 337, 357,  60,  70, 287, 223, 354,  16, 254,
            267,  72,  99, 192, 191, 346,  49, 146, 289,  29, 120, 328, 183,
            134,   9,  27, 221, 356, 110, 124]
```

```python
train_element_counts = {}
test_element_counts = {}

for i, file_num in enumerate(file_data['document'].keys()):
    file = '../../thesis_data/bpmai/models/' + file_num + '.meta.json'
    with open(file, 'r') as f:
        data = f.read()
        meta_data = json.loads(data)
        element_counts = meta_data['revision']['elementCounts']
        
    for k, v in element_counts.items():
        if re.search('(Task|Event|Gateway|SequenceFlow)', k) :
            if i in test_idx:
                if k in test_element_counts.keys():
                    test_element_counts[k].append(v)
                else:
                    test_element_counts.update({k: [v]})       
            else: 
                if k in train_element_counts.keys():
                    train_element_counts[k].append(v)
                else:
                    train_element_counts.update({k: [v]})
```

```python
print(len(train_element_counts.keys()))
print(len(test_element_counts.keys()))
```

```python
train_df = pd.DataFrame.from_dict(train_element_counts, orient='index')
test_df = pd.DataFrame.from_dict(test_element_counts, orient='index')
test_df.head()
```

```python
train_df = train_df.T
test_df = test_df.T
test_df.head()
```

```python
# set a grey background (use sns.set_theme() if seaborn version 0.11.0 or above) 
sns.set(style="darkgrid")

sns.histplot(data=train_df, x="Task", color="skyblue", label="train_Task", kde=True)
sns.histplot(data=test_df, x="Task", color="red", label="test_Task", kde=True)

plt.legend() 
plt.show()
```
